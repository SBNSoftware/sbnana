```python
def makeplot(dict, variable, labels, detVar_labels,
             myrange=(0,180), xlabel = 'x axis', 
             bins = 25, plotname_to_save = 'plotname',
             density = True, logy = False,
             ratio_bounds=None):
    for benchmark in dict:
        fig, (p0, p1) = plt.subplots(2, 1, gridspec_kw={'height_ratios': [3,1]}, sharex=True) #, figsize=(7, 6.4)
        fig.subplots_adjust(hspace=0.03)
        cv_hist = 0 # placeholder
        bin_centers = 0 # placeholder
        for v, vl in enumerate(variation_labels):
            df = dict[benchmark][v]
            x = p0.hist(df[variable]*180./math.pi, weights = df.scale,
                     bins=bins, label=vl, histtype='step',
                     linewidth=2, range=myrange, density=density)
            if v==0: 
                cv_hist = x
                bin_centers = 0.5*(x[1][1:] + x[1][:-1])
            p1.errorbar( bin_centers, x[0]/cv_hist[0])
        
        # p0 formatting:
        if density:
            p0.set_ylabel("Area Normalized")
        else:
            p0.set_ylabel("Events / %s" % POTSTR)
        if logy: plt.yscale("log")
        p0.text(0.4, 0.96, benchmark, fontsize=14,
            transform=p0.transAxes, verticalalignment="top")
        
        # p1 formatting:
        p1.set_xlabel(xlabel)
        p1.set_ylabel("Det. Var. / CV")
        if ratio_bounds is not None:
            p1.set_ylim(ratio_bounds)
        else:
            p1.set_ylim([0,2])
        
        # common formatting:
        plt.xlim(myrange)
        #legend
        box = p0.get_position()
        p0.set_position([box.x0, box.y0, box.width * 0.8, box.height])
        p0.legend(loc='lower center', bbox_to_anchor=(0.5, 1.03), ncol=3, fontsize='x-small')
        box = p1.get_position()
        p1.set_position([box.x0, box.y0, box.width * 0.8, box.height])
        
        plt.show()
```
